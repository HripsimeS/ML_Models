## Exercise 8
1. Using the while() function, create a loop that calculates the factorial of a number.
2. Calculate the factorial of n = 10. 
3. Compare this with the result obtained with the predefined function on R “factorial(n)”.


```r
#1
fac = function(n){
  i = 2
  fact = 1
  while(i<=n){
    fact = fact * i
    i = i + 1
  }
  fact
}
```

```r
#2, 3
fac(10)
factorial(10)
```

1. Create a function which for a given pair $(n, p)$ with natural $n$ and $p$ in $[0, 1]$, evaluates the maximum of the error made when we approach the Binomial law by Poisson law $$M_{n,p} = \max|P(X_n = k) - P(Y_n = k)|$$
with $k = 0, \dots n$, 
Where $X_n$ follows the Binomial Law, of parameter (n, p) and Yn follows the Poisson Law of parameter np.
2.	For $p = 1/2$, draw the graph of the error according to $n$.
3.	For $n = 40$, draw the graph of the error according to $p$.
4. 	Comment on the results obtained.



```r
#1

get_error= function(n,p){
  values = (1:n)
  lambda =n*p
  liste =abs(dpois(values,lambda)-dbinom(values,n,p))
  return(max(abs(dpois(values,lambda)-dbinom(values,n,p))))
}
get_error(10,0.5)
```

```r
#2

N =(1:1000)
p = 0.5
errors = sapply(N,get_error, p=0.5)
par(mfrow = c(1,2))
plot(N,errors,pch =20, col = 'red', main ="Variation of error depending on n ",xlab = "Error", ylab = "n")
lines(N,errors)
```

```r
#3
P = seq(0,1,0.01)
errors = sapply(P,get_error, n=40)
plot(P,errors,pch =20, col = 'red', main ="Variation of error depending on p ",xlab = "Error", ylab = "p")
lines(P,errors)
```

We notice that the approaches to 0 when n is important with p fixed. We have also that the error approaches 0 for low values of p with n fixed. This comes from the fact that a binomial distribution converges to a Poisson distribution n goes to infinity with np fixed or at least p tends has low enough value.
