```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue ='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue ='sex', split = True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue ='sex', jitter=True)
```

```python
sns.violinplot (x='day', y='total_bill', data=tips, color = 'pink')
sns.swarmplot(x='day', y='total_bill', data=tips, hue ='sex', palette=['green', 'blue'])
```

```python
sns.lmplot(x='total_bill', y= 'tip', data =tips, hue ='sex')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.kdeplot)
g.map_lower(plt.scatter)
```

```python
f = sns.FacetGrid(tips, col='time', row= 'smoker', hue='sex')
f.map(sns.distplot, 'total_bill')
```
