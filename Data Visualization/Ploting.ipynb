```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,20)
y=x**2
```

```python
plt.plot(x,y)
plt.xlabel('This X Axis')
plt.ylabel('This Y Axis')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='red')
plt.subplot(1,2,2)
plt.plot(y,x,color='blue')
```

```python
fig = plt.figure()
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig =plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.3,0.3,0.4,0.3])
axes1.plot(x,y, color='red')
axes2.plot(y,x, color='green')
axes1.set_title('First Plot')
axes.set_title('Seco Plot')
```
