```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_context('paper')
sns.set_style('whitegrid')
from matplotlib import style
style.use('dark_background')
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=500, n_features=2, centers=4, cluster_std=1.8, random_state=42)
```

```python
plt.scatter(data[0][:,0], data[0][:, 1], c=data[1], cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
plt.scatter(data[0][:,0], data[0][:, 1], c=kmeans.labels_, cmap='rainbow')
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10, 6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:, 1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:, 1], c=data[1], cmap='rainbow')
```

```python
X, y_true = make_blobs(n_samples=1500, n_features=3, centers=5, cluster_std=1.5, random_state=42)
```

```python
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:,0], X[:, 1],X[:,2], c=y_true)
```

```python
plt.figure(figsize=(15,7))
plt.scatter(X[:,0], X[:, 1], c=y_true, cmap='rainbow')
```

```python
plt.figure(figsize=(15,7))
plt.scatter(X[:,1], X[:, 2], c=y_true, cmap='rainbow')
```

```python
plt.figure(figsize=(15,7))
plt.scatter(X[:,0], X[:, 2], c=y_true, cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, axes = plt.subplots(3,2, sharey=True, figsize=(18, 18))

axes[0][0].set_title('K Means')
axes[0][0].scatter(X[:,0], X[:, 1], c=kmeans.labels_, cmap='rainbow')

axes[0][1].set_title('Original')
axes[0][1].scatter(X[:,0], X[:, 1], c=y_true, cmap='rainbow')

axes[1][0].set_title('K Means')
axes[1][0].scatter(X[:,1], X[:, 2], c=kmeans.labels_, cmap='rainbow')

axes[1][1].set_title('Original')
axes[1][1].scatter(X[:,1], X[:, 2], c=y_true, cmap='rainbow')

axes[2][0].set_title('K Means')
axes[2][0].scatter(X[:,0], X[:, 2], c=kmeans.labels_, cmap='rainbow')

axes[2][1].set_title('Original')
axes[2][1].scatter(X[:,0], X[:, 2], c=y_true, cmap='rainbow')
```
