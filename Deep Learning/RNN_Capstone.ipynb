```python
test_day= 2
test_ind=test_day*25
```

```python
test_ind
```

```python
# Notice the minus sign in our indexing
train = df.iloc[:-test_ind]
test = df.iloc[-test_ind:]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(train)
```

```python
scaled_train = scaler.transform(train)
scaled_test = scaler.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
# define generator
length = 25 # Length of the output sequences (in number of timesteps)
batch_size = 1 #Number of timeseries samples in each batch
generator = TimeseriesGenerator(scaled_train, scaled_train, length=length, batch_size=batch_size)
```

```python
len(scaled_train)
```

```python
len(generator) 
```

```python
# What does the first batch look like?
X,y = generator[0]
```

```python
print(f'Given the Array: \n{X.flatten()}')
print(f'Predict this y: \n {y}')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM
```

```python
scaled_train.shape
```

```python
# define model
model = Sequential()

# Simple RNN layer
model.add(LSTM(300, activation='relu', input_shape=(length,scaled_train.shape[1])))

# Final Prediction (one neuron per feature)
model.add(Dense(scaled_train.shape[1]))

model.compile(optimizer='adam', loss='mse')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss',patience=1)
validation_generator = TimeseriesGenerator(scaled_test,scaled_test, 
                                           length=length, batch_size=batch_size)
```

```python
model.fit_generator(generator,epochs=10,
                    validation_data=validation_generator,
                   callbacks=[early_stop])
```

```python
model.history.history.keys()
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
n_features = scaled_train.shape[1]
test_predictions = []

first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(len(test)):
    
    # get prediction 1 time stamp ahead ([0] is for grabbing just the number instead of [array])
    current_pred = model.predict(current_batch)[0]
    
    # store prediction
    test_predictions.append(current_pred) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
test_predictions
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
```

```python
true_predictions
```

```python
true_predictions = pd.DataFrame(data=true_predictions,columns=test.columns)
```

```python
true_predictions.head()
```

```python
np.sqrt(mean_squared_error(true_predictions['traffic_volume'], test['traffic_volume']))
```
