```python
import h5py
```

```python
with h5py.File('Datasets/MNIST/Mini_data.hdf5') as f:
    X_train = f['X'][:]
    y_train = f['y'][:]
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
import tensorflow as tf
```

```python
X = tf.placeholder(tf.float32, shape=[100, 784])
y = tf.placeholder(tf.float32, shape=[100, 10])
```

```python
W1 = tf.Variable(tf.random_normal(shape=[784, 256], stddev=0.01))
b1 = tf.Variable(tf.zeros(shape=[256]))

W2 = tf.Variable(tf.random_normal(shape=[256, 10], stddev=0.01))
b2 = tf.Variable(tf.zeros(shape=[10]))
```

```python
Z1 = tf.add(tf.matmul(X,W1),b1)
A1 = tf.tanh(Z1)

out = tf.add(tf.matmul(A1,W2),b2)
```

```python
# lost = -(y*tf.log(A) + (1-y)*tf.log(1-A))
lost = tf.nn.softmax_cross_entropy_with_logits_v2(labels=y, logits=out)
cost = tf.reduce_mean(lost)
GradientDescent =  tf.train.GradientDescentOptimizer(learning_rate=0.01)
train_op = GradientDescent.minimize(cost)
```

```python
correct_pred = tf.equal(tf.argmax(out,1), tf.argmax(y,1))
accuracy = tf.reduce_mean(tf.cast(correct_pred,tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoch in range(10000):
        acc, _ = sess.run([accuracy, train_op], {X:X_train, y:y_train})
        if epoch%10 == 0:
            print('Epoch', epoch, 'Accuracy', acc)
```
