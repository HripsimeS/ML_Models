```python
import h5py
```

```python
with h5py.File('Mini_MNIST.hdf5') as f:
    X_train = f['X'][:]
    y_train = f['y'][:]
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
import tensorflow as tf
```

```python
X = tf.placeholder(tf.float32, shape=[100, 784])
y = tf.placeholder(tf.int16, shape=[100, 10])
```

```python
W = tf.Variable(tf.random_normal(shape=[784, 10]))
b = tf.Variable(tf.zeros(shape=[10]))
```

```python
Z = tf.matmul(X,W) + b
```

```python
lost = tf.nn.softmax_cross_entropy_with_logits_v2(labels=y, logits=Z)
cost = tf.reduce_mean(lost)
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train_op = optimizer.minimize(cost) #training operation
```

```python
correct_prediction = tf.equal(tf.arg_max(Z,1), tf.arg_max(y,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoch in range(20000):
        _, acc, c = sess.run([train_op, accuracy, cost], {X:X_train, y:y_train})
        if epoch %10 ==0:
            print("Epoch", epoch, "Accuracy", acc, "Cost", c)
```
